```python
import numpy as np
import pandas as pd
```

```python
# the second name is in Mongo
la_liga_teams = [
    ('Real Madrid','Real Madrid CF'), 
    ('Atlético Madrid','Atlético de Madrid'), 
    ('Real Sociedad','Real Sociedad'), 
    ('Sevilla','Sevilla FC'), 
    ('Betis','Real Betis Balompié'), 
    ('Rayo Vallecano','Rayo Vallecano'),
    ('Barcelona','FC Barcelona'),
    ('Athletic Club','Athletic Club de Bilbao'),
    ('Espanyol','RCD Espanyol de Barcelona'),
    ('Osasuna','CA Osasuna'),
    ('Valencia','Valencia CF'),
    ('Villarreal','Villarreal CF'),
    ('Celta Vigo','RC Celta de Vigo'),
    ('Mallorca','RCD Mallorca'),
    ('Alavés','Deportivo Alavés'),
    ('Granada','Granada CF'),
    ('Elche','Elche CF'),
    ('Cádiz','Cádiz CF'),
    ('Getafe','Getafe CF'),
    ('Levante','Levante Unión Deportiva'),
]

teams = [team for _, team in la_liga_teams]
```

```python
data = pd.read_csv('result4.csv')
```

```python
players = []

for player_name in np.unique(data.player):

    la_liga_player = data[data.player == player_name]

    la_liga_player = la_liga_player[la_liga_player['Team'].isin(teams)]

    
    t2 = []
    
    for team_name, salary in zip(np.array(la_liga_player['Team']), np.array(la_liga_player['prediction'])):
        t = {}
        t['team_name'] = team_name
        t['prediction'] = salary
        
        t2.append(t)
        
    players.append({'name': player_name, 'predictions': t2})
    
    if (len(players) % 100) == 0: 
#         break
        print(len(players))
```

```python
import json

with open("predictions.json", "w", encoding='utf16') as outfile:
    
    json.dump({'model':players}, outfile)
```
